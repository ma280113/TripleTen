#### **Ejercicio 1:** Almacenar nombres en una lista

- Input: Lista de diccionarios.
- Output: Lista de strings.

  1. Inicializar una lista vacia.

  4. lista.append(nombre)

````python
personas = [
    {"nombre": "Ana", "edad": 25},
    {"nombre": "Luis", "edad": 30},
    {"nombre": "María", "edad": 28}
]

# Inicializar la lista para almacenar nombres
nombres = []


for persona in personas:
  nombre = persona['nombre']
  print(persona)
# Instrucción: Almacenar el nombre de cada persona en la lista
````

#### **Ejercicio 2:** Calcular la suma de edades
Usando la misma lista de diccionarios, utiliza un bucle for para calcular la suma de las edades de todas las personas y almacenar el resultado en una variable.

- input -> lista de diccionarios
- output -> numero entero- > suma de las edades

proceso:

1. generar una variable = 0
2. definir el ciclo for. v_temporal = dict.
3. extraer la edad -> ['llave']
4. Sumar la edad

````python
personas = [
    {"nombre": "Ana", "edad": 25},
    {"nombre": "Luis", "edad": 30},
    {"nombre": "María", "edad": 28}
]

# Inicializar la variable para la suma
suma_edades = 0

for persona in personas:
  edad = persona['edad']
  suma_edades += edad
print(suma_edades)
# Instrucción: Calcular la suma de las edades
````

#### **Ejercicio 3:** Almacenar información de cada persona en una lista
Utiliza un bucle for para almacenar la información de cada persona en una nueva lista en el siguiente formato: "Nombre: Ana, Edad: 25, Ciudad: Madrid".

````python
personas = [
    {"nombre": "Ana", "edad": 25, "ciudad": "Madrid"},
    {"nombre": "Luis", "edad": 30, "ciudad": "Barcelona"},
    {"nombre": "María", "edad": 28, "ciudad": "Sevilla"}
]
````

Salida esperada:

```python

]
```

### Respuestas

#### Ejercicio 1: Almacenar nombres en una lista

````python
personas = [
    {"nombre": "Ana", "edad": 25},
    {"nombre": "Luis", "edad": 30},
    {"nombre": "María", "edad": 28}
]

# Inicializar la lista para almacenar nombres
nombres = []

# Almacenar el nombre de cada persona en la lista
for persona in personas:
    nombres.append(persona["nombre"])

# Resultado esperado: ["Ana", "Luis", "María"]
print(nombres)
````

#### Ejercicio 2: Calcular la suma de edades

````python
# Inicializar la variable para la suma
suma_edades = 0

# Calcular la suma de las edades
for persona in personas:
    suma_edades += persona["edad"]

# Resultado esperado: 83
print(suma_edades)
````

#### Ejercicio 5: Almacenar información de cada persona en una lista

````python
personas = [
    {"nombre": "Ana", "edad": 25, "ciudad": "Madrid"},
    {"nombre": "Luis", "edad": 30, "ciudad": "Barcelona"},
    {"nombre": "María", "edad": 28, "ciudad": "Sevilla"}
]

# Inicializar la lista para almacenar la información
informacion_personas = []

# Almacenar la información de cada persona en la lista
for persona in personas:
    info = f"Nombre: {persona['nombre']}, Edad: {persona['edad']}, Ciudad: {persona['ciudad']}"
    informacion_personas.append(info)

print(informacion_personas)
````
