## Ejercicios

### 1. Contar letras en una palabra
Escribe un bucle for que recorra cada letra en la palabra "Pythonista" y cuente cuántas veces aparece la letra "t".

### 2. Suma acumulativa hasta 50

Utiliza un bucle while para sumar números del 1 en adelante hasta que la suma total supere 50. Al final, imprime el último número que se sumó y el total.

### 3.Filtrar números pares
Dada la lista de números numeros = [3, 8, 15, 24, 7, 13, 18], usa un bucle for para crear una nueva lista que contenga solo los números pares.

### 4. Identificar palabras que contienen "a"
Dada una lista de palabras palabras = ["gato", "perro", "avión", "cielo", "árbol"], usa un bucle for e imprime solo las palabras que contienen la letra "a".

### 5.Clasificar estudiantes por notas
Ejercicio: Dado un diccionario con las notas de estudiantes notas = {"Ana": 85, "Luis": 76, "Carlos": 90, "Marta": 65}, escribe un bucle que clasifique e imprima si cada estudiante "aprobó" (nota >= 70) o "reprobó".

### 6. Conteo por tipo de numero.

Dada una lista de números numeros = [4, -1, 0, 7, -3, 5, 0, -8], utiliza un bucle for y condiciones lógicas para contar cuántos de estos números son positivos, negativos y ceros. Almacena los resultados en un diccionario, de manera que cada tipo de número (positivo, negativo o cero) solo cree una clave en el diccionario si hay al menos uno de ese tipo en la lista.

## Respuestas

### 1. Contar letras en una palabra

```python
# Palabra a analizar
palabra = "Pythonista"

# Contador de letras 't'
contador = 0

# Bucle que recorre cada letra
for letra in palabra:
    if letra == "t":
        contador += 1

print("La letra 't' aparece", contador, "veces en la palabra", palabra)
```

### 2. Suma acumulativa hasta 50

```python
# Inicialización de variables
suma_total = 0
numero = 1

# Bucle que suma números hasta que la suma total supere 50
while suma_total <= 50:
    suma_total += numero
    numero += 1

print("La suma total es:", suma_total, "y el último número sumado fue:", numero - 1)
```

### 3. Filtrar numeros pares

```python
# Lista de números
numeros = [3, 8, 15, 24, 7, 13, 18]

# Lista para almacenar los números pares
numeros_pares = []

# Bucle para filtrar los números pares
for numero in numeros:
    if numero % 2 == 0:
        numeros_pares.append(numero)

print("Números pares:", numeros_pares)
```

### 4. Identificar palabras que contienen la "a"

```python
# Lista de palabras
palabras = ["gato", "perro", "avión", "cielo", "árbol"]

# Bucle para identificar palabras con "a"
for palabra in palabras:
    if "a" in palabra:
        print(palabra)
```

### 5. Clasificar estudiantes por notas

```python
# Diccionario de notas
notas = {"Ana": 85, "Luis": 76, "Carlos": 90, "Marta": 65}

# Bucle para clasificar estudiantes
for estudiante, nota in notas.items():
    if nota >= 70:
        print(estudiante, "aprobó con una nota de", nota)
    else:
        print(estudiante, "reprobó con una nota de", nota)
```

### 6. Conteo por tipo de numero.

```python
# Lista de números
numeros = [4, -1, 0, 7, -3, 5, 0, -8]

# Diccionario para almacenar los conteos
conteo = {}

# Bucle para clasificar cada número y actualizar el diccionario
for numero in numeros:
    if numero > 0:
        if "positivos" not in conteo:
            conteo["positivos"] = 0
        conteo["positivos"] += 1
    elif numero < 0:
        if "negativos" not in conteo:
            conteo["negativos"] = 0
        conteo["negativos"] += 1
    else:
        if "ceros" not in conteo:
            conteo["ceros"] = 0
        conteo["ceros"] += 1

print(conteo)
```
