## Ejercicio 5: Contar empleados en cada departamento

Crea una función llamada `contar_empleados` que reciba una lista de diccionarios anidados. Cada diccionario representa un departamento con un nombre y una lista de empleados. La función debe devolver un nuevo diccionario donde las claves son los nombres de los departamentos y los valores son el número de empleados en cada uno.

**Ejemplo de entrada**:
```python
departamentos = [
    {"departamento": "Recursos Humanos", "empleados": ["Juan", "Ana"]},

    {"departamento": "Ventas", "empleados": ["Sara"]}

# Respuestas

## Ejercicio 1: Calcular la suma total de precios

````python
def calcular_total(productos):
    total = 0
    for producto in productos:
        if producto["precio"] > 0:  # Solo suma precios positivos
            total += producto["precio"]
    return total

# Ejemplo de uso
productos = [
    {"nombre": "Producto A", "precio": 10},
    {"nombre": "Producto B", "precio": -5},
    {"nombre": "Producto C", "precio": 15}
]

print("Total:", calcular_total(productos))  # Salida esperada: 25
````

## Ejercicio 2: Encontrar estudiantes aprobados

````python
def estudiantes_aprobados(estudiantes):
    aprobados = []
    for estudiante in estudiantes:
        promedio = sum(estudiante["calificaciones"]) / len(estudiante["calificaciones"])
        if promedio >= 6:
            aprobados.append(estudiante["nombre"])
    return aprobados

# Ejemplo de uso
estudiantes = [
    {"nombre": "Ana", "calificaciones": [5, 6, 7]},
    {"nombre": "Luis", "calificaciones": [4, 5, 5]},
    {"nombre": "Carla", "calificaciones": [8, 7, 9]}
]

print("Estudiantes aprobados:", estudiantes_aprobados(estudiantes))  # Salida esperada: ["Ana", "Carla"]
````

## Ejercicio 3: Contar palabras en una lista de listas

````python
def contar_palabras(palabras):
    conteo = {}
    for sublista in palabras:
        for palabra in sublista:
            if palabra in conteo:
                conteo[palabra] += 1
            else:
                conteo[palabra] = 1
    return conteo

# Ejemplo de uso
palabras = [["hola", "mundo"], ["hola", "python"], ["python", "mundo"]]
print("Conteo de palabras:", contar_palabras(palabras))  # Salida esperada: {"hola": 2, "mundo": 2, "python": 2}
````

## Ejercicio 4: Buscar productos en una lista anidada

````python
def buscar_producto(tiendas, producto_buscado):
    for tienda in tiendas:
        if producto_buscado in tienda["productos"]:
            return tienda["tienda"]
    return "Producto no encontrado"

# Ejemplo de uso
tiendas = [
    {"tienda": "Tienda 1", "productos": ["pan", "leche", "huevos"]},
    {"tienda": "Tienda 2", "productos": ["queso", "yogurt"]},
    {"tienda": "Tienda 3", "productos": ["leche", "mantequilla"]}
]

print(buscar_producto(tiendas, "leche"))  # Salida esperada: "Tienda 1"
print(buscar_producto(tiendas, "carne"))  # Salida esperada: "Producto no encontrado"
````

## Ejercicio 5: Contar empleados en cada departamento

````python
def contar_empleados(departamentos):
    conteo = {}
    for depto in departamentos:
        conteo[depto["departamento"]] = len(depto["empleados"])
    return conteo

# Ejemplo de uso
departamentos = [
    {"departamento": "Recursos Humanos", "empleados": ["Juan", "Ana"]},
    {"departamento": "TI", "empleados": ["Carlos", "Luis", "Pedro"]},
    {"departamento": "Ventas", "empleados": ["Sara"]}
]

print("Conteo de empleados por departamento:", contar_empleados(departamentos))  # Salida esperada: {"Recursos Humanos": 2, "TI": 3, "Ventas": 1}
````
