# ¿Qué es una función en Python?

Una **función** en Python es un bloque de código reutilizable que realiza una tarea específica. Las funciones nos ayudan a organizar y simplificar el código, permitiendo que escribamos una vez un conjunto de instrucciones y lo utilicemos varias veces.

Imagina que una función es como una **receta de cocina**: defines los pasos una vez (al escribir la función), y luego puedes "seguir la receta" cada vez que necesites, simplemente llamando a la función.

#### ¿Por qué usar funciones?

Las funciones tienen varias ventajas:
- **Reutilización**: Puedes llamar a la misma función varias veces sin reescribir el código.
- **Organización**: Ayudan a dividir el código en secciones más pequeñas y manejables.
- **Facilitan el mantenimiento**: Si necesitas hacer un cambio, solo lo haces en la función y se reflejará en todas las partes donde la usas.

#### ¿Cómo se define una función?

Para definir una función en Python, usamos la palabra clave `def` seguida del nombre de la función y paréntesis `()`:

```python
def nombre_de_la_funcion():
    # Aquí van las instrucciones de la función
    print("Esta es una función de ejemplo")

# Partes de una función en Python

Una función en Python tiene varias partes importantes. Conocerlas te ayudará a estructurar y entender cómo funcionan.

## 1. Palabra clave `def`

Toda función comienza con la palabra clave `def`, que le indica a Python que estás creando una nueva función.

```python
def
```

## 2.Nombre de la función
A continuación, le das un nombre a la función. El nombre debe describir su propósito y seguir las reglas de nombres en Python (sin espacios, sin caracteres especiales, y no puede comenzar con un número). Es común usar snake_case (todo en minúsculas, palabras separadas por guiones bajos).

```python
def nombre_de_la_funcion
```

## 3. Parametros ()
Después del nombre, se colocan paréntesis. Dentro de estos paréntesis puedes definir los parámetros de la función, que son las variables que la función usará al ser llamada. Si no necesita parámetros, los paréntesis quedan vacíos.

```python
def nombre_de_la_funcion(parametro1, parametro2)
```

## 4. Dos puntos :
El símbolo : al final indica que las instrucciones de la función vienen a continuación.

```python
def nombre_de_la_funcion(parametro1, parametro2):
```

## 5. Cuerpo de la función (instrucciones)
El cuerpo de la función es el bloque de código indentado que contiene las instrucciones de la función. Estas instrucciones se ejecutarán cuando la función sea llamada.

```python
def nombre_de_la_funcion(parametro1, parametro2):
    # Aquí van las instrucciones
    resultado = parametro1 + parametro2
    print("La suma es:", resultado)
```

## 6. return (opcional)
La instrucción return permite devolver un valor desde la función. No es obligatorio, pero si la función necesita dar un resultado, return lo devuelve y la función se detiene en ese punto.

```python
def sumar(a, b):
    resultado = a + b
    return resultado  # devuelve el valor de "resultado"
```

## Función para dar la bienvenida a una lista de personas

La siguiente función llamada `bienvenida_grupal` toma como parámetro una lista de nombres y, para cada nombre en la lista, imprime un mensaje de bienvenida. No usa `return` porque no necesita devolver un valor; solo realiza una acción (imprimir mensajes).

```python
def bienvenida_grupal(nombres):
    for nombre in nombres:
        print("¡Bienvenido/a, " + nombre + "! Nos alegra que estés aquí.")

# Llamada de ejemplo
lista_de_nombres = ["Ana", "Luis", "Pedro", "Maria"]
bienvenida_grupal(lista_de_nombres)
```

Explicación del código:

- Parámetro: La función bienvenida_grupal recibe un parámetro nombres, que se espera que sea una lista de nombres.
- Bucle for: Dentro de la función, se usa un bucle for para recorrer cada nombre en la lista nombres.
- Acción (sin return): Para cada nombre, la función imprime un mensaje de bienvenida personalizado. No es necesario devolver ningún valor, ya que la función solo imprime los mensajes.

input: lista de strings.
output: lista de saludos, lista de strings.

Receta cocina:
1. Crear la variable de lista de saludos = []
2. Iterar sobre la lista ( ciclo for)
3. Agregar el nombre del cliente al saludo.
3. Agregar el anterior a la lista final

````python
lista_de_nombres_1 = ["Ana", "Luis", "Pedro", "Maria"]  # semana 1
lista_de_nombres_2 = ["Anas", "Luiss", "Pedros", "Marias"] # semana 2

def bienvenida_grupal(nombres):
    lista_acum = []
    for nombre in nombres:
      # lista_acum.append("¡Bienvenid@ " + nombre + "! Nos alegra que estés aquí.")
      lista_acum.append(f"¡Bienvenid@ {nombre}! Nos alegra que estés aquí.")
    print(lista_acum)

variable_1 = bienvenida_grupal(lista_de_nombres_1)
````

## Ejemplo: Función para calcular el promedio de una lista de números

La siguiente función, llamada `calcular_promedio`, toma como parámetro una lista de números, calcula su promedio y devuelve el valor promedio usando `return`.

```python
def calcular_promedio(numeros):
    suma = sum(numeros)  # Suma todos los elementos de la lista
    cantidad = len(numeros)  # Cuenta el número de elementos en la lista
    promedio = suma / cantidad  # Calcula el promedio
    return promedio  # Devuelve el promedio

# Llamada de ejemplo
lista_numeros = [10, 20, 30, 40, 50]
resultado = calcular_promedio(lista_numeros)
print("El promedio es:", resultado)
```
