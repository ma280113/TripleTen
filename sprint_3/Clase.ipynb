# üìö ¬øQu√© es pandas?
Pandas es una librer√≠a de Python dise√±ada espec√≠ficamente para trabajar con datos de manera eficiente y sencilla. Su principal objetivo es proporcionar herramientas para:

- Organizar datos en estructuras como tablas, muy similares a las hojas de c√°lculo de Excel.
- Manipular y analizar datos de forma flexible.
- Cargar y guardar datos desde archivos como CSV, Excel, bases de datos, entre otros.

# üåü Caracter√≠sticas clave de pandas:
Estructuras principales:
- Series: Una columna de datos (como una lista con etiquetas).
- DataFrame: Una tabla de datos con filas y columnas (similar a una hoja de c√°lculo).
- F√°cil manejo de datos faltantes.
- Potentes herramientas para filtrar, agrupar y transformar datos.
- Compatibilidad con otras librer√≠as de Python para ciencia de datos, como NumPy y Matplotlib.

# üß† ¬øPor qu√© es importante?

Pandas es fundamental para cualquier an√°lisis de datos porque te permite transformar datos desordenados en informaci√≥n clara y procesable, sin necesidad de conocimientos avanzados en programaci√≥n.

# üìö ¬øQu√© es un DataFrame?
Un DataFrame es la estructura principal de datos que ofrece la librer√≠a pandas. Es como una tabla de datos, similar a las hojas de c√°lculo de Excel o tablas en bases de datos.

üß† Caracter√≠sticas clave:
- Forma tabular: Los datos est√°n organizados en filas y columnas.
- Etiquetas:
  - Filas: Cada fila tiene un √≠ndice (por defecto, un n√∫mero entero).
  - Columnas: Cada columna tiene un nombre o etiqueta.
- Versatilidad: Puede contener diferentes tipos de datos en columnas (n√∫meros, texto, fechas, etc.).
- Operaciones avanzadas: Puedes filtrar, agregar, transformar, y analizar datos de forma r√°pida y eficiente.


üõ†Ô∏è ¬øDe d√≥nde vienen los datos en un DataFrame?
Un DataFrame se puede crear a partir de:
- Diccionarios.
- Listas de listas.
- Archivos (CSV, Excel, JSON).
- Bases de datos.

# üìö ¬øQu√© es una Serie?
Una Serie en pandas es una estructura de datos que se asemeja a una columna en una hoja de c√°lculo o en una tabla. Es un arreglo unidimensional que puede contener datos de cualquier tipo, como n√∫meros, texto, fechas, entre otros.

üß† Caracter√≠sticas clave:

- Unidimensional: Es como una lista o un array, pero con funcionalidades adicionales.
- √çndice asociado: Cada elemento de la Serie tiene una etiqueta √∫nica (llamada √≠ndice) que permite identificar los valores de manera flexible.
Versatilidad: Soporta operaciones avanzadas como filtrado, agregaci√≥n, y m√°s.

üåü Diferencia entre una Serie y un DataFrame:
- Una Serie es una sola columna.
- Un DataFrame es una tabla completa (varias columnas y filas).

![Series](https://www.tomasbeuzen.com/python-programming-for-data-science/_images/series.png)


https://www.tomasbeuzen.com/python-programming-for-data-science/chapters/chapter7-pandas.html


````python
# Crear una Serie a partir de una lista
edades = pd.Series(
    [23, 30, 25],
    index=["Ana", "Luis", "Carlos"]
  )
edades
````

````python
edades > 28
````

````python
edades[edades > 28]
````

````python

# Mostrar la Serie
print("Serie creada:")
display(edades)
print("---------------------")

# Acceder a un elemento por su √≠ndice
print("\nEdad de Luis:")
display(edades["Luis"])
print("---------------------")

# Operaciones con Series
print("\nEdades mayores de 25:")
display(edades[edades > 25])
print("---------------------")

# Convertir una Serie en una lista
edades_lista = edades.tolist()
print("\nSerie convertida en lista:")
display(edades_lista)
print("---------------------")

# Convertir una Serie en un diccionario
edades_diccionario = edades.to_dict()
print("\nSerie convertida en diccionario:")
display(edades_diccionario)
print("---------------------")

# Mostrar los primeros registros
print("\nAcceder a los primeros 2 registros:")
display(edades.head(2))
print("---------------------")

# Mostrar los ultimos registros
print("\nAcceder a los ultimos 2 registros:")
display(edades.tail(2))
print("---------------------")

# Acceder al indice
print("\nMostrar los indices:")
display(edades.index)
print("---------------------")

# Acceder a los valores
print("\nMostrar los valores:")
display(edades.values)
print("---------------------")
````

# üìö Cargar archivos CSV y Excel con pandas
La librer√≠a pandas incluye funciones f√°ciles de usar para cargar datos desde archivos comunes como CSV y Excel directamente a un DataFrame.

## üß† Funciones principales:
- Cargar archivos CSV: Usamos pd.read_csv().
- Cargar archivos Excel: Usamos pd.read_excel().

  Ambas funciones transforman los datos del archivo en un DataFrame, que es la estructura principal de pandas para trabajar con datos tabulares.

üõ†Ô∏è Cargar un archivo CSV
  Un archivo CSV (Comma-Separated Values) contiene datos separados por comas. Pandas puede leerlo f√°cilmente.

## üìå Eliminaci√≥n de duplicados en Pandas con `drop_duplicates()`

Cuando trabajamos con datos en **Pandas**, a veces encontramos filas duplicadas que pueden afectar nuestros an√°lisis. La funci√≥n `drop_duplicates()` nos ayuda a eliminar esas filas repetidas de manera r√°pida y sencilla.

### üßê ¬øCu√°ndo usar `drop_duplicates()`?
- Cuando tienes datos con filas repetidas y solo necesitas valores √∫nicos.
- Para limpiar un DataFrame antes de un an√°lisis o visualizaci√≥n.
- Para asegurarte de que no haya registros duplicados en bases de datos.

### üîß Par√°metros principales:
- `subset`: Especifica las columnas en las que queremos buscar duplicados.
- `keep`: Controla qu√© duplicados se eliminan:
  - `"first"` (predeterminado): Mantiene la primera aparici√≥n y elimina las siguientes.
  - `"last"`: Mantiene la √∫ltima aparici√≥n y elimina las anteriores.
  - `False`: Elimina todas las filas duplicadas.
- `inplace`: Si es `True`, modifica el DataFrame original en lugar de devolver uno nuevo.

### üéØ Ejemplo de uso:
Imagina que tenemos una lista de clientes con registros duplicados y queremos limpiarla.


````python
# Crear un DataFrame con valores duplicados
data = {
    'Nombre': ['Ana', 'Luis', 'Ana', 'Carlos', 'Luis', 'Pedro','Ana'],
    'Edad': [25, 30, 25, 40, 30, 35, 50],
    'Ciudad': ['Medellin', 'Medell√≠n', 'Medellin', 'Cali', 'Medell√≠n', 'Cartagena','Cali']
}

df = pd.DataFrame(data)
df
````

````python
df.drop_duplicates()
````

````python
df
````

````python
df.drop_duplicates(subset = ['Nombre'], keep = 'last')
````

````python
df = df.drop_duplicates()

df.drop_duplicates(inplace = True)
````

````python
df
````

````python


print("üìå DataFrame original:")
print(df)

# Eliminar duplicados basados en todas las columnas
df_sin_duplicados = df.drop_duplicates()

print("\n‚úÖ DataFrame sin duplicados:")
print(df_sin_duplicados)

# Eliminar duplicados basados solo en la columna 'Nombre'
df_sin_duplicados_nombre = df.drop_duplicates(subset=['Nombre'])

print("\nüéØ DataFrame sin duplicados en 'Nombre':")
print(df_sin_duplicados_nombre)
````

# üìö ¬øQu√© es groupby?
groupby se utiliza para:

- Agrupar datos seg√∫n los valores de una o m√°s columnas.
- Aplicar funciones de agregaci√≥n a los grupos (como sum, mean, count, etc.).

## üß† Proceso de groupby:
1. Dividir los datos en grupos seg√∫n los valores de una o m√°s columnas.
2. Aplicar una operaci√≥n (como sumar o promediar) a cada grupo.
3. Combinar los resultados en un nuevo DataFrame o Serie.

````python
import pandas as pd

# Crear un DataFrame con datos de ventas
data = {
    "Vendedor": ["Ana", "Luis", "Ana", "Carlos", "Luis", "Carlos", "Ana"],
    "Producto": ["A", "B", "A", "C", "B", "C", "C"],
    "Ventas": [100, 200, 150, 300, 250, 400, 120]
}

df = pd.DataFrame(data)

print("DataFrame inicial:")
df
````

````python
df.groupby(['Vendedor','Producto'])['Ventas'].mean()
````

## üõ†Ô∏è Usos comunes de groupby
1. Suma de ventas por vendedor

````python
# Agrupar por "Vendedor" y sumar las ventas
# ventas_por_vendedor = df.groupby(["Vendedor",'Producto'], as_index = False)["Ventas"].mean()

ventas_por_vendedor = df.groupby(["Vendedor",'Producto'])["Ventas"].mean().reset_index()

print("\nVentas totales por vendedor:")
ventas_por_vendedor
````

````python
# Agrupar por "Vendedor" y sumar las ventas
ventas_por_vendedor = df.groupby(["Vendedor",'Producto'])["Ventas"].mean()

print("\nVentas totales por vendedor:")
ventas_por_vendedor
````
