### Ejercicio 1: Creación y selección en DataFrames

Crea un DataFrame utilizando `pandas` con la siguiente información de ventas:

| Producto  | Cantidad | Precio |
|-----------|----------|--------|
| Manzanas  | 10       | 2.0    |
| Naranjas  | 5        | 1.5    |
| Plátanos  | 8        | 0.8    |
| Manzanas  | 15       | 2.1    |
| Naranjas  | 3        | 1.4    |

1. Selecciona y muestra solo las filas donde el producto sea `"Manzanas"`.
2. Selecciona y muestra únicamente las columnas `"Producto"` y `"Cantidad"`.

````python
import pandas as pd

# Creación del DataFrame
data = {
    "Producto": ["Manzanas", "Naranjas", "Plátanos", "Manzanas", "Naranjas"],
    "Cantidad": [10, 5, 8, 15, 3],
    "Precio": [2.0, 1.5, 0.8, 2.1, 1.4],
}
ventas = pd.DataFrame(data)

# Selección de las filas donde el producto sea "Manzanas"
manzanas = ventas.loc[ventas["Producto"] == "Manzanas"]
print("Filas donde el producto es 'Manzanas':")
print(manzanas)

# Selección de las columnas "Producto" y "Cantidad"
producto_cantidad = ventas.loc[:, ["Producto", "Cantidad"]]
print("\nColumnas 'Producto' y 'Cantidad':")
print(producto_cantidad)
````

### Ejercicio 2: Series y estadísticas básicas

Usando el DataFrame del ejercicio anterior:

1. Extrae la columna `"Cantidad"` como una `Series`.
2. Calcula y muestra:
   - La cantidad total vendida.
   - El promedio de cantidad vendida por producto.
   - La cantidad máxima y mínima.

````python
# Extracción de la columna "Cantidad" como una Series
cantidad_series = ventas["Cantidad"]

# Cálculos estadísticos
total_vendido = cantidad_series.sum()
promedio_vendido = cantidad_series.mean()
cantidad_max = cantidad_series.max()
cantidad_min = cantidad_series.min()

print(f"Total vendido: {total_vendido}")
print(f"Promedio vendido: {promedio_vendido:.2f}")
print(f"Máxima cantidad: {cantidad_max}")
print(f"Mínima cantidad: {cantidad_min}")
````

### Ejercicio 3: Uso de `.loc` con condiciones múltiples

Usando el DataFrame del ejercicio 1:

1. Selecciona las filas donde la cantidad vendida sea mayor a 5 y el precio sea menor a 2.
2. Modifica los valores de `"Cantidad"` a 20 para las filas donde el producto sea `"Plátanos"`.
3. Muestra el DataFrame resultante.

````python
# Selección con condiciones múltiples
condiciones = (ventas["Cantidad"] > 5) & (ventas["Precio"] < 2)
seleccion = ventas.loc[condiciones]
print("Filas con cantidad > 5 y precio < 2:")
print(seleccion)

# Modificación de los valores donde el producto es "Plátanos"
ventas.loc[ventas["Producto"] == "Plátanos", "Cantidad"] = 20
print("\nDataFrame después de modificar 'Plátanos':")
print(ventas)
````

### Ejercicio 4: Agrupación con `groupby`

Usando el DataFrame inicial:

1. Agrupa los datos por `"Producto"` y calcula:
   - La suma total de `"Cantidad"` por producto.
   - El precio promedio por producto.
2. Muestra los resultados.

````python
# Agrupación por producto y cálculos
agrupado = ventas.groupby("Producto").agg(
    Total_Cantidad=("Cantidad", "sum"),
    Promedio_Precio=("Precio", "mean"),
)
print("Resultados de la agrupación:")
print(agrupado)
````

### Ejercicio 5: Agrupación por dos columnas

Usando el siguiente DataFrame de ventas:

| Producto  | Región   | Cantidad | Precio |
|-----------|----------|----------|--------|
| Manzanas  | Norte    | 10       | 2.0    |
| Naranjas  | Norte    | 5        | 1.5    |
| Plátanos  | Sur      | 8        | 0.8    |
| Manzanas  | Norte    | 15       | 2.1    |
| Naranjas  | Sur      | 3        | 1.4    |
| Manzanas  | Sur      | 7        | 2.0    |
| Plátanos  | Norte    | 12       | 0.9    |

1. Agrupa los datos por las columnas `"Producto"` y `"Región"`.
2. Calcula:
   - La cantidad total vendida para cada grupo.
   - El precio promedio para cada grupo.
3. Muestra los resultados en un nuevo DataFrame.

```python

data = {
    "Producto": ["Manzanas", "Naranjas", "Plátanos", "Manzanas", "Naranjas", "Manzanas", "Plátanos"],
    "Región": ["Norte", "Norte", "Sur", "Norte", "Sur", "Sur", "Norte"],
    "Cantidad": [10, 5, 8, 15, 3, 7, 12],
    "Precio": [2.0, 1.5, 0.8, 2.1, 1.4, 2.0, 0.9],
}
ventas = pd.DataFrame(data)

```

````python
# Agrupación por Producto y Región
agrupado = ventas.groupby(["Producto", "Región"]).agg(
    Total_Cantidad=("Cantidad", "sum"),
    Promedio_Precio=("Precio", "mean"),
)

# Mostrar los resultados
print("Resultados de la agrupación por Producto y Región:")
print(agrupado)
````
